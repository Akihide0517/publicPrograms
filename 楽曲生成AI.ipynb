```python
# ライブラリのインポート
from mingus.containers import Note, Track, Composition, Bar
from mingus.midi import midi_file_out
import subprocess
```

```python
# 音源を作成する関数
def generate_song(song_notes, notes_lengths, filename, meter):
  # トラック作成
  track = Track()

  # 新しい小節を作成して各音符を追加
  bar = Bar()
  bar.meter = meter  #曲の拍子
  for note, length in zip(song_notes, notes_lengths):
    if note == "rest":
      if not bar.place_rest(length):
        track.add_bar(bar)
        bar = Bar()
        bar.meter = meter
        bar.place_rest(length)
    else:
      if not bar.place_notes(note, length):
        # 小節をトラックに追加
        track.add_bar(bar)
        bar = Bar()
        bar.meter = meter
        bar.place_notes(note, length)

  # 最後の小節を追加
  track.add_bar(bar)

  # 楽曲にトラックを追加
  composition = Composition()
  composition.add_track(track)

  # MIDIファイルを生成
  midi_filename = filename + ".mid"
  midi_file_out.write_Composition(midi_filename, composition)

  # MIDIをWAVに変換
  wav_filename = filename + ".wav"
  subprocess.run(["fluidsynth", "-ni", "FluidR3_GM.sf2", midi_filename, "-F", wav_filename, "-r", "44100"])
```

楽曲作成

```python
# ライブラリのインポート
import pretty_midi
import numpy as np
import os
from google.colab import files
```

```python
# MIDIファイルをアップロード
uploaded = files.upload()
```

```python
# データセット保持リスト
datasets = []

# アップロードされた各MIDIファイルに対して処理
for midi_file in uploaded.keys():
    if midi_file.endswith('.mid') or midi_file.endswith('.midi'):
        # MIDIデータの読み込み
        midi_data = pretty_midi.PrettyMIDI(midi_file)

        all_notes = []

        for instrument in midi_data.instruments:
            for note in instrument.notes:
                # 各音の開始時間、終了時間、音の高さを学習の特徴量として使用
                all_notes.append([note.start, note.end, note.pitch])

        # 数値化
        all_notes_array = np.array(all_notes)

        # 正規化
        max_values = np.max(all_notes_array, axis=0)
        min_values = np.min(all_notes_array, axis=0)
        normalized_all_notes = (all_notes_array - min_values) / (max_values - min_values)

        # 系列データの作成
        sequence_length = 10
        X = []
        y = []
        for i in range(len(normalized_all_notes) - sequence_length):
            X.append(normalized_all_notes[i:i + sequence_length])
            y.append(normalized_all_notes[i + sequence_length])
        X = np.array(X)
        y = np.array(y)

        # この楽曲のデータセットをリストに追加
        datasets.append((X, y))
```

学習

```python
# ライブラリのインポート
from keras.models import Sequential
from keras.layers import LSTM, Dense
import random
```

```python
# LSTMモデルの設計
model = Sequential()
# 入力形状：10タイムステップ、3特徴量
model.add(LSTM(128, input_shape=(10, 3)))
# 全結合層の追加
model.add(Dense(3))
```

```python
# モデルのコンパイル
model.compile(optimizer='adam', loss="mse")
```

```python
# エポック数
epochs = 50

# 学習ループ
for epoch in range(epochs):
    X_batch, y_batch = random.choice(datasets)
    model.train_on_batch(X_batch, y_batch)
```

```python
# データセットの一部を入力として使用
input_data = X_batch[0].reshape(1, 10, 3)

# ここでは、30音の楽曲を生成。song_lengthの値を変更してお好きな長さの曲を生成してください。
song_length = 30
generated_music = []
for i in range(song_length):
    predicted_note = model.predict(input_data)
    generated_music.append(predicted_note[0])
    input_data = np.append(input_data[:, 1:, :], predicted_note.reshape(1, 1, 3), axis=1)
```

```python
# 逆正規化
denormalized_music = generated_music * (max_values - min_values) + min_values

# MIDIオブジェクトの作成
generated_midi = pretty_midi.PrettyMIDI()

# instrumentを作成
instrument_program = pretty_midi.instrument_name_to_program('Acoustic Grand Piano')
instrument = pretty_midi.Instrument(program=instrument_program)

# 開始時間
current_time = 0.0
for note_array in denormalized_music:
    start, end, pitch = note_array
    time_interval = end - start

    start = current_time
    end = current_time + time_interval
    note = pretty_midi.Note(velocity=127, pitch=int(pitch), start=start, end=end)
    instrument.notes.append(note)

    current_time += time_interval

# MIDIファイルへの書き出し
generated_midi.instruments.append(instrument)
output_file_path = "generated_music.mid"
generated_midi.write(output_file_path)
```
